```python
# М*А*Т=К+Р+У+Ж+О+К
from collections import defaultdict

def is_unique3(s):
    dd = defaultdict(bool)
    for char in s:
        if dd[char]:
            return False
        dd[char] = True
    return True

max_d = 8
s = set([n for n in range(1, max_d + 1)])

iter_cnt, cnt = 0, 0
for m in s:
    for a in s - set([m]):
        for t in s - set([m, a]):
            for k in s - set([m, a, t]):
                for r in s - set([m, a, t, k]):
                    for y in s - set([m, a, t, k, r]):
                        for j in s - set([m, a, t, k, r, y]):
                            for o in s - set([m, a, t, k, r, y, j]):
                                iter_cnt += 1
                                s1 = ''.join(map(str, [m, a, t, k, r, y, j, o]))
                                if (m * a * t == k + r + y + j + o + k) and is_unique3(s1):
                                    cnt += 1
                                    print(f'{m} * {a} * {t} = {k} + {r} + {y} + {j} + {o} + {k}')
print(f'Общее количество решений: {cnt}')
print(f'Количество итераций: {iter_cnt}')
```
